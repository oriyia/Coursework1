```python
import pandas as pd
# import dask.dataframe as dd

# импорт всех необходимых csv-файлов
students = pd.read_csv('students.csv', parse_dates=['birthday'])
courses = pd.read_csv('courses.csv').drop('Unnamed: 0', axis=1)
courses.columns = ['id', 'course_title', 'course_field']
course_contents = pd.read_csv('course_contents.csv')
progresses = pd.read_csv('progresses.csv')
progress_phases = pd.read_csv('progress_phases.csv')
```

```python
# создание одного общего датасета
df = pd.merge(progress_phases, progresses, right_on='id', left_on='progress_id', how='left').drop('id', axis=1)
df = pd.merge(df, courses, left_on='course_id', right_on='id', how='left').drop('id', axis=1)
df = pd.merge(df,
              course_contents[['course_id', 'module_title', 'lesson_title', 'lesson_token', 'is_video', 'is_homework',
                               'module_number', 'lesson_number']],
              left_on=['course_id', 'module_number', 'lesson_number'],
              right_on=['course_id', 'module_number', 'lesson_number'],
              how='inner',
              suffixes=('', '_y'))
df.drop(df.filter(regex='_y$').columns.tolist(), axis=1, inplace=True)
```

```python
# датасет для расчета среднего возраста студентов
students_birthday = students.dropna(subset=['birthday']).reset_index(drop=True)
```

**1.2 Количество модулей в каждом курсе:**

```python
df.groupby('course_title')['module_title'].nunique().reset_index().sort_values(by='module_title', ascending=False)
```

**1.3 Количество уроков в каждом модуле на каждом курсе:**

```python
df_lesson_count = df.groupby(['course_title', 'module_title'])['lesson_title'].nunique().reset_index()
df_lesson_count.sort_values(by='lesson_title', ascending=False)
```

**1.4 Медианное количество уроков в модуле на каждом курсе:**

```python
df_lesson_count.groupby(['course_title'])['lesson_title'].aggregate(['median']).sort_values(by='median', 
                                                                                            ascending=False).reset_index()
```

**1.5 Количество учеников на каждом курсе**

```python
count_students = df.groupby('course_title')['student_id'].nunique().reset_index().sort_values(by='student_id', ascending=False)
count_students
```

**1.6 Минимальный, максимальный, средний, медианный возраст студентов:**

```python
import datetime

# принудительно приводим столбец "birthday" к формату datetime (пропущенные значения=NaT)
students_birthday['birthday'] = pd.to_datetime(students_birthday['birthday'], errors='coerce')

# устанавливаем ограничение по возрасту, так как есть аномальные значения (человек не может быть возраста: 1 год, 100 и 
# более лет). Пусть будет от 12 до 100 лет.
students_birthday_red = students_birthday[(students_birthday['birthday'] > '1920') & (students_birthday['birthday'] <= '2008')]
# вычисляем возраст, как разницу между сегодняшней датой и датой рождения, результат получается в днях, делим на 365, 
# получаем год. Результат сохраняем в новый стлобец "age".
students_birthday_red['age'] = (datetime.datetime.now() - students_birthday_red['birthday']).apply(lambda x: (x / 365).days)
students_birthday_red['age'].agg(['min', 'max', 'mean', 'median']).reset_index()
```

**1.7 Минимальный, максимальный, средний, медианный возраст студентов на каждом курсе**

```python
df_age = pd.merge(df[['student_id', 'course_title']], students_birthday_red[['id', 'age']], 
                  left_on='student_id', right_on='id', how='inner')
# удаляем дубликаты строк
df_age = df_age.drop_duplicates(subset=['student_id', 'course_title'])
```

```python
df_age.groupby('course_title')['age'].agg(['min', 'max', 'mean', 'median'])
```
